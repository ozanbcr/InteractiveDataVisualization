```python
mapped = df_world.set_index('name').join(df.set_index('birth_country')).reset_index()
dfmap=mapped.rename(columns={"index" : "country"})
import plotly.express as px

fig = px.choropleth(dfmap, locations="iso_a3",
                    color="nobel_count",
                    hover_name="country",color_continuous_scale="Viridis"
                    )
fig.update_layout(
    title_text='Nobel Prize per Country',
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'))
fig.show()
```

```python
gender_df = df.groupby(['prize_year','gender'])["prize"].count().reset_index(name = 'Nobel Prize Count')
fig2=px.bar(gender_df, x="prize_year",y="Nobel Prize Count",color="gender",labels={ "gender": "Gender/Organization"})
fig2.update_layout(
    title_text='Annual Nobel Prize Distribution')


fig2.show()
```

**Şekil 4**

```python
fig3=px.pie(gender_df,values="Nobel Prize Count",names="gender")
fig3.update_layout(
    title_text='Nobel Prize Distribution')
fig3.show()
```
